```python
data_path = './../../ASCERTAIN_Features/Dt_SelfReports.mat'

self_reports = scipy.io.loadmat(data_path)
self_reports = self_reports['Ratings']
```

```python
data_frames = []

for user_id in range(1, 59):
    print('User ID: ' + str(user_id))
    
    presentation_id = []
    
    for clip_id in range(1, 37):
        
        presentation_id.append(str(user_id).zfill(2) + str(clip_id).zfill(2))
        
    user_data = pd.DataFrame({
        'presentation_id' : presentation_id,
        'arousal' : self_reports[0, user_id - 1, :],
        'valence' : self_reports[1, user_id - 1, :],
        'engaging' : self_reports[2, user_id - 1, :],
        'liking' : self_reports[3, user_id - 1, :],
        'familiarity' : self_reports[4, user_id - 1, :]
    })
        
    data_frames.append(user_data)
```

```python
Ratings = pd.concat(data_frames,ignore_index=True)
```
